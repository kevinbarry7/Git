```python
from sklearn.linear_model import LinearRegression
regr = linear_model.LinearRegression()
results = {}
for feature in features:    
    X = crime[feature].values.reshape(-1, 1)
    results[feature] = {}
    #print(x_train)
    for response in responses:
        y = crime[response].values
        results[feature][response] = []
        for i in range(len(y)):
            y_val = y[i]
            if math.isnan(y_val) or y_val == float('infinity') or y_val == float('-infinity'):
                y[i] = 0
                #print("y", i, "set to zero from bad value ", y_val)
        regr.fit(X, y)
        rss = np.mean((regr.predict(X) - y) ** 2)
        variance_score = regr.score(X, y)
        results[feature][response].extend([regr.coef_, rss, variance_score])
        
#print(results)
        
# The coefficients
#print('Coefficients: \n', regr.coef_)
# The mean square error
#print("Residual sum of squares: %.2f"
#      % np.mean((regr.predict(X) - y) ** 2))
# Explained variance score: 1 is perfect prediction
#print('Variance score: %.2f' % regr.score(X, y))

# Print X vs y and prediction.
#plt.scatter(X, y, color='black')
#plt.plot(X, regr.predict(X), color='blue', linewidth=3)
#plt.show()
```
